```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

def plot_gsb(v, elev, azim, roll):
  # f4 = [ [2,3,5,4], [6,8,7,9] ]
  fa = [ [0,2,3], [0,3,5], [0,5,4], [0,4,2] ]
  fb = [ [1,6,8], [1,8,7], [1,7,9], [1,9,6] ]
  fc = [ [2,4,8], [2,8,6], [2,3,6], [3,6,9], [3,9,5], [5,7,9], [4,8,7], [5,4,7] ]

  f = fc + fb + fa  # 16 faces # f4 + fc + fb + fa
  verts =  [ [ v[i] for i in p] for p in f]  # list comprehension
  colors = ['red','blue','green','orange','violet','aqua','magenta','teal'] + ['indianred', 'burlywood', 'grey', 'coral', 'indianred', 'burlywood', 'grey', 'coral'] # 'slateblue', 'dodgerblue', 'yellowgreen', 'pink']

  fig = plt.figure(figsize=plt.figaspect(1))
  ax = fig.add_subplot(projection='3d')
  ax.set(xlim=(-1.0,1.0), ylim=(-1.0,1.0), zlim=(-1.0,1.0))
  ax.view_init(elev, azim, roll)
  ax.add_collection3d(Poly3DCollection(verts, edgecolors='k', facecolors='w', linewidths=1, alpha=0.8))
  fig.tight_layout()
  plt.show()
```

```python
v = [
[ 0.96242838, -0.16512661,  0.56377749],
[-0.96242838,  0.16512661, -0.56377749],
[ 0.53565169,  0.61520348,  0.10667071],
[ 0.03565169,  0.11520348,  0.81377749],
[ 0.6820983,  -0.23834991, -0.39332929],
[ 0.1820983,  -0.73834991,  0.31377749],
[-0.46242838,  0.66512661,  0.14332929],
[-0.2553216,  -0.54198018, -0.56377749],
[-0.0053216,   0.41512661, -0.7102241],
[-0.71242838, -0.29198018,  0.2897759] # 10 vertices
]

plot_gsb(v, 45, 110, 0)
```
